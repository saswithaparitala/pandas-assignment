```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/drive/MyDrive/winequality - winequality.csv')
count=df['quality'].value_counts()
count_list=[]
qualitylabels=df.quality.unique()
for i in qualitylabels:
  count_list.append(count[i])
plt.bar(qualitylabels,count_list)
```
